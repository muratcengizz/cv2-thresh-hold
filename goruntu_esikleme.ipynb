```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(filename='img1.jpg')
img = cv2.cvtColor(src=img, code=cv2.COLOR_BGR2GRAY)
plt.figure()
plt.imshow(X=img, cmap='gray');
plt.axis('off')
plt.show()
```

```python
retval, thresh_img = cv2.threshold(src=img, thresh=60, maxval=255, type=cv2.THRESH_BINARY)
```

```python
plt.figure()
plt.imshow(X=thresh_img, cmap='gray')
plt.show()
```

```python
retval, thresh_img2 = cv2.threshold(src=img, thresh=60, maxval=255, type=cv2.THRESH_BINARY_INV)
plt.figure()
plt.imshow(X=thresh_img2, cmap='gray')
plt.show()
```

```python
thresh_image3 = cv2.adaptiveThreshold(src=img, 
                                      maxValue=255, 
                                      adaptiveMethod=cv2.ADAPTIVE_THRESH_MEAN_C,
                                      thresholdType=cv2.THRESH_BINARY,
                                      blockSize=11,
                                      C=8)
plt.figure()
plt.imshow(X=thresh_image3, cmap='gray')
plt.show()
```

## cv2.threshold(src, thresh, maxval, type)
### src: İşlenecek olan giriş görüntüsüdür. Bu bir Numpy dizisi olarak temsil edilir.
### thresh: Eşik değeridir. Giriş görüntüsündeki piksel değerleri, bu eşik değeriyle karşılaştırılır ve belirli bir kurala göre ikili görüntüye dönüştürülür.
### maxval: Maksimum piksel değeri, ikili görüntüye dönüştürülürken kullanılır. Eğer bir piksel değeri eşik değerinden büyükse, bu maksimum değer atanır; aksi halde, sıfır atanır.
### type: Eşikleme yöntemini belirten bir bayrak veya yöntem türüdür. Bu parametre, görüntüyü ikili hale dönüştürmek için kullanılan spesifik eşikleme yöntemini kontrol eder. Örneğin, 'cv2.THRESH_BINARY' ikili eşikleme yöntemini temsil eder.

### Bu fonksiyon, iki değer döndürür: retval: Kullanılan eşik değerini döndürür. dst: İkili görüntüyü döndürür. Bu da bir Numpy dizisi olarak temsil edilir.

## cv2.adaptiveThreshold(src, maxValue, adaptiveMethod, thresholdType, blockSize, C)
### src: Giriş görüntüsüdür. Bu, eşikleme uygulanacak olan tek kanallı bir görüntü olmalıdır.
### maxValue: Eşikleme sonucunda alınabilecek maksimum değerdir. Eşikleme sonucu, bu değeri geçen piksellerin değeri maxValue olarak ayarlanacaktır. 
### adaptiveMethod: Uygulanacak adaptif eşikleme yöntemini belirler. Bu parametre, cv2.ADAPTIVE_THRESH_MEAN_C veya cv2.ADAPTIVE_THRESH_GAUSSIAN_C olarak ayarlanabilir. İlk yöntem, piksellerin etrafındaki bir bölgeye ortalamayı alarak eşik değerini belirlerken, ikinci yöntem piksellerin etrafındaki bir bölgeye Gauss filtresi uygulayarak eşik değerini belirler.
### thresholdType: Eşikleme türünü belirler. Bu, cv2.THRESH_BINARY veya cv2.THRESH_BINARY_INV olarak ayarlanabilir. İlk türde, eşik değerini geçen pikseller maxValue olarak ayarlanırken, ikinci türde değeri geçen pikseller 0 olarak ayarlanır.
### blockSize: Adaptif eşikleme için kullanılacak piksel bloğunun boyutunu belirler. Bu, tek bir eşik değeri hesaplamak için kullanılan piksel bloğunun boyutunu belirtir. Örneğin, blockSize=3, 3x3 piksel blokların kullanılacağını gösterir.
### C: Adaptif eşikleme için eşik değeri düzeltmesini belirler. Bu değer, hesaplanan eşik değerine eklenir veya çıkarılır. Eşik değeri = ortalamaDeğer - C.
