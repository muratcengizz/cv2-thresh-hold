```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(filename='img1.jpg')
img = cv2.cvtColor(src=img, code=cv2.COLOR_BGR2GRAY)
plt.figure()
plt.imshow(X=img, cmap='gray');
plt.axis('off')
plt.show()
```

```python
retval, thresh_img = cv2.threshold(src=img, thresh=60, maxval=255, type=cv2.THRESH_BINARY)
```

```python
plt.figure()
plt.imshow(X=thresh_img, cmap='gray')
plt.show()
```

```python
retval, thresh_img2 = cv2.threshold(src=img, thresh=60, maxval=255, type=cv2.THRESH_BINARY_INV)
plt.figure()
plt.imshow(X=thresh_img2, cmap='gray')
plt.show()
```

```python
thresh_image3 = cv2.adaptiveThreshold(src=img, 
                                      maxValue=255, 
                                      adaptiveMethod=cv2.ADAPTIVE_THRESH_MEAN_C,
                                      thresholdType=cv2.THRESH_BINARY,
                                      blockSize=11,
                                      C=8)
plt.figure()
plt.imshow(X=thresh_image3, cmap='gray')
plt.show()
```

## cv2.threshold(src, thresh, maxval, type)
### src: lenecek olan giri grntsdr. Bu bir Numpy dizisi olarak temsil edilir.
### thresh: Eik deeridir. Giri grntsndeki piksel deerleri, bu eik deeriyle karlatrlr ve belirli bir kurala gre ikili grntye dntrlr.
### maxval: Maksimum piksel deeri, ikili grntye dntrlrken kullanlr. Eer bir piksel deeri eik deerinden bykse, bu maksimum deer atanr; aksi halde, sfr atanr.
### type: Eikleme yntemini belirten bir bayrak veya yntem trdr. Bu parametre, grnty ikili hale dntrmek iin kullanlan spesifik eikleme yntemini kontrol eder. rnein, 'cv2.THRESH_BINARY' ikili eikleme yntemini temsil eder.

### Bu fonksiyon, iki deer dndrr: retval: Kullanlan eik deerini dndrr. dst: kili grnty dndrr. Bu da bir Numpy dizisi olarak temsil edilir.

## cv2.adaptiveThreshold(src, maxValue, adaptiveMethod, thresholdType, blockSize, C)
### src: Giri grntsdr. Bu, eikleme uygulanacak olan tek kanall bir grnt olmaldr.
### maxValue: Eikleme sonucunda alnabilecek maksimum deerdir. Eikleme sonucu, bu deeri geen piksellerin deeri maxValue olarak ayarlanacaktr. 
### adaptiveMethod: Uygulanacak adaptif eikleme yntemini belirler. Bu parametre, cv2.ADAPTIVE_THRESH_MEAN_C veya cv2.ADAPTIVE_THRESH_GAUSSIAN_C olarak ayarlanabilir. lk yntem, piksellerin etrafndaki bir blgeye ortalamay alarak eik deerini belirlerken, ikinci yntem piksellerin etrafndaki bir blgeye Gauss filtresi uygulayarak eik deerini belirler.
### thresholdType: Eikleme trn belirler. Bu, cv2.THRESH_BINARY veya cv2.THRESH_BINARY_INV olarak ayarlanabilir. lk trde, eik deerini geen pikseller maxValue olarak ayarlanrken, ikinci trde deeri geen pikseller 0 olarak ayarlanr.
### blockSize: Adaptif eikleme iin kullanlacak piksel blounun boyutunu belirler. Bu, tek bir eik deeri hesaplamak iin kullanlan piksel blounun boyutunu belirtir. rnein, blockSize=3, 3x3 piksel bloklarn kullanlacan gsterir.
### C: Adaptif eikleme iin eik deeri dzeltmesini belirler. Bu deer, hesaplanan eik deerine eklenir veya karlr. Eik deeri = ortalamaDeer - C.
